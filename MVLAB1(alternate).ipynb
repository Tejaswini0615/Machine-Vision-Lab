```python
import numpy as np
from PIL import Image
```

```python
# Load images as NumPy arrays
color_img1 = np.array(Image.open("/home/student/Downloads/garden-rose-red-pink-56866.jpeg"))
color_img2 = np.array(Image.open("/home/student/Downloads/kitten.jpg"))
gray_img1 = np.array(Image.open("/home/student/Downloads/garden-rose-red-pink-56866.jpeg").convert("L"))
gray_img2 = np.array(Image.open("/home/student/Downloads/kitten.jpg").convert("L"))
```

```python
# Resize images if needed
color_img2 = np.resize(color_img2, color_img1.shape)
gray_img2 = np.resize(gray_img2, gray_img1.shape)
```

```python
# Arithmetic Operations
add_color = np.clip(color_img1 + color_img2, 0, 255).astype(np.uint8)
subtract_color = np.clip(color_img1 - color_img2, 0, 255).astype(np.uint8)
multiply_color = np.clip(color_img1 * color_img2, 0, 255).astype(np.uint8)
divide_color = np.clip(color_img1 / (color_img2 + 1e-5), 0, 255).astype(np.uint8)
```

```python
# Logical Operations
and_gray = np.bitwise_and(gray_img1, gray_img2)
or_gray = np.bitwise_or(gray_img1, gray_img2)
xor_gray = np.bitwise_xor(gray_img1, gray_img2)
not_gray1 = np.bitwise_not(gray_img1)
```

```python
# Convert results to Image format and display
Image.fromarray(add_color).show("Added Color")
Image.fromarray(subtract_color).show("Subtracted Color")
Image.fromarray(multiply_color).show("Multiplied Color")
Image.fromarray(divide_color).show("Divided Color")
Image.fromarray(and_gray).show("AND Grayscale")
Image.fromarray(or_gray).show("OR Grayscale")
Image.fromarray(xor_gray).show("XOR Grayscale")
Image.fromarray(not_gray1).show("NOT Grayscale")
```

```python
# 3. Histogram Analysis

# Compute and display the histogram for a grayscale image.
# Compute and display the color histograms (R, G, B) for a color image.
# Perform histogram equalization to enhance the contrast.
```

```python
# With OpenCV
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load grayscale and color images
gray_image = cv2.imread("/home/student/Downloads/garden-rose-red-pink-56866.jpeg", cv2.IMREAD_GRAYSCALE)
color_image = cv2.imread("/home/student/Downloads/garden-rose-red-pink-56866.jpeg")

# 1. Compute and display the histogram for a grayscale image
gray_hist = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.plot(gray_hist)
plt.xlim([0, 256])
plt.show()

# 2. Compute and display the color histograms (R, G, B) for a color image
colors = ('b', 'g', 'r')  # OpenCV uses BGR
plt.figure()
plt.title("Color Histograms")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
for i, color in enumerate(colors):
    color_hist = cv2.calcHist([color_image], [i], None, [256], [0, 256])
    plt.plot(color_hist, color=color)
    plt.xlim([0, 256])
plt.show()

# 3. Perform histogram equalization to enhance contrast
equalized_gray = cv2.equalizeHist(gray_image)

# For color images, apply histogram equalization to each channel separately in YUV color space
yuv_image = cv2.cvtColor(color_image, cv2.COLOR_BGR2YUV)
yuv_image[:, :, 0] = cv2.equalizeHist(yuv_image[:, :, 0])
equalized_color = cv2.cvtColor(yuv_image, cv2.COLOR_YUV2BGR)

# Display results
cv2.imshow("Original Grayscale", gray_image)
cv2.imshow("Equalized Grayscale", equalized_gray)
cv2.imshow("Original Color", color_image)
cv2.imshow("Equalized Color", equalized_color)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
# Load grayscale and color images
gray_image = Image.open("/home/student/Downloads/garden-rose-red-pink-56866.jpeg").convert("L")  # Convert to grayscale
color_image = Image.open("/home/student/Downloads/garden-rose-red-pink-56866.jpeg")  # Color image
```

```python
# Convert images to NumPy arrays
gray_array = np.array(gray_image)
color_array = np.array(color_image)
```

```python
# 1. Compute and display the histogram for a grayscale image
gray_hist, bins = np.histogram(gray_array.flatten(), bins=256, range=[0, 256])

plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.plot(gray_hist)
plt.xlim([0, 256])
plt.show()
```

```python
# 2. Compute and display the color histograms (R, G, B) for a color image
colors = ('r', 'g', 'b')  # PIL uses RGB
plt.figure()
plt.title("Color Histograms")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
for i, color in enumerate(colors):
    color_hist, _ = np.histogram(color_array[:, :, i].flatten(), bins=256, range=[0, 256])
    plt.plot(color_hist, color=color)
    plt.xlim([0, 256])
plt.show()
```

```python
# 3. Perform histogram equalization to enhance contrast
# Grayscale equalization
gray_cdf = np.cumsum(np.histogram(gray_array.flatten(), bins=256, range=[0, 256])[0])
gray_cdf_normalized = (gray_cdf - gray_cdf.min()) * 255 / (gray_cdf.max() - gray_cdf.min())
equalized_gray = np.interp(gray_array.flatten(), range(256), gray_cdf_normalized).reshape(gray_array.shape).astype(np.uint8)
```

```python
# Color equalization: Apply to each channel
equalized_color = np.zeros_like(color_array)
for i in range(3):  # R, G, B
    channel = color_array[:, :, i]
    channel_cdf = np.cumsum(np.histogram(channel.flatten(), bins=256, range=[0, 256])[0])
    channel_cdf_normalized = (channel_cdf - channel_cdf.min()) * 255 / (channel_cdf.max() - channel_cdf.min())
    equalized_color[:, :, i] = np.interp(channel.flatten(), range(256), channel_cdf_normalized).reshape(channel.shape).astype(np.uint8)
```

```python
# Convert equalized grayscale and color back to images
equalized_gray_image = Image.fromarray(equalized_gray)
equalized_color_image = Image.fromarray(equalized_color)
```

```python
# Display results
equalized_gray_image.show("Equalized Grayscale")
equalized_color_image.show("Equalized Color")
```
